```python
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
import numpy as np
```

Let's generate our dataset and take a look on it.

```python
moons_points, moons_labels = make_moons(n_samples=500, noise=0.2, random_state=42)
plt.scatter(moons_points[:, 0], moons_points[:, 1], c=moons_labels)
```

## 1.1 Pure models.
First let's try to solve this case with good old Logistic Regression and simple (linear kernel) SVM classifier.

Train LR and SVM classifiers (choose params by hand, no CV or intensive grid search neeeded) and plot their decision regions. Calculate one preffered classification metric.

_Tip:_ to plot classifiers decisions you colud use either sklearn examples ([this](https://scikit-learn.org/stable/auto_examples/neural_networks/plot_mlp_alpha.html#sphx-glr-auto-examples-neural-networks-plot-mlp-alpha-py) or any other) and mess with matplotlib yourself or great [mlxtend](https://github.com/rasbt/mlxtend) package (see their examples for details)

_Pro Tip:_ wirte function `plot_decisions` taking a dataset and an estimator and plotting the results cause you want to use it several times below

```python
# Разделим данные на test и train выборки
from sklearn.model_selection import train_test_split

train_moons_points, test_moons_points, train_moons_labels, test_moons_labels = train_test_split(
    moons_points, moons_labels, test_size=0.3, random_state=42)
```

```python
def plot_decision_regs(X, y, estimator, poly=None, cmap=ListedColormap(['#a9c7d6','#FFAFAF'])):
    fig = plt.figure(figsize=(8, 6))
    x_min = X[:, 0].min() - 0.5
    x_max = X[:, 0].max() + 0.5
    y_min = X[:, 1].min() - 0.5
    y_max = X[:, 1].max() + 0.5
    
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 1000), np.linspace(y_min, y_max, 1000))
    features = None
    
    if poly is not None:
        features = poly.fit_transform(np.c_[xx.ravel(), yy.ravel()])
    else: 
        features = np.c_[xx.ravel(), yy.ravel()]
        
    preds = estimator.predict(features)
    preds = preds.reshape(xx.shape)
    plt.pcolormesh(xx, yy, preds, cmap=cmap)
    plt.scatter(X[:, 0], X[:, 1], c=y)
    plt.grid()
    plt.show()
```

```python
# Обучим линейную регрессию
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score
from matplotlib.colors import ListedColormap


lr = LogisticRegression(random_state=42, penalty='l2', solver='sag', max_iter=1000, 
                            tol=0.001, C=0.3)
lr.fit(train_moons_points, train_moons_labels)
predictions = lr.predict(test_moons_points)

print('F1-score of logistic regression classifier on test set: {:.2f}'.format(
    f1_score(y_true=test_moons_labels, y_pred=predictions)))
```

```python
# Решающая поверхность для линейной регрессии
plot_decision_regs(moons_points, moons_labels, lr)
```

```python
# Обучим SVM
from sklearn.svm import SVC

svm = SVC(kernel='linear', C=0.1, tol=0.01)

svm.fit(train_moons_points, train_moons_labels)
predictions = svm.predict(test_moons_points)

print('F1-score of SVM classifier on test set: {:.2f}'.format(
    f1_score(y_true=test_moons_labels, y_pred=predictions)))
```

```python
# Решающая поверхность для SVM
plot_decision_regs(moons_points, moons_labels, svm)
```

`Conclusion:` из графиков видно, что линейные классификаторы недостаточно хорошо разделяют данные сложной структуры. Ожидается, что SVM с нелинейными ядрами будет работать лучше. 

## 1.2 Kernel tirck

![keep-calm](imgs/keep-calm.jpg)

Now use different kernels (`poly`, `rbf`, `sigmoid`) on SVC to get better results. Play `degree` parameter and others.

For each kernel estimate optimal params, plot decision regions, calculate metric you've chosen eariler.

Write couple of sentences on:

```python
# Polynomial kernel('poly')

svm_poly = SVC(kernel='poly', C=5.0, degree=3.0)
svm_poly.fit(train_moons_points, train_moons_labels)
predictions = svm_poly.predict(test_moons_points)

print('F1-score of SVM classifier with polynomial kernel function on test set: {:.2f}'.format(
    f1_score(y_true=test_moons_labels, y_pred=predictions)))

plot_decision_regs(moons_points, moons_labels, svm_poly)
```

```python
# RBF kernel

svm_rbf = SVC(kernel='rbf', C=30.0)
svm_rbf.fit(train_moons_points, train_moons_labels)
predictions = svm_rbf.predict(test_moons_points)

print('F1-score of SVM classifier with RBF kernel function on test set: {:.2f}'.format(
    f1_score(y_true=test_moons_labels, y_pred=predictions)))

plot_decision_regs(moons_points, moons_labels, svm_rbf)
```

```python
# Sigmoid kernel 

svm_sigmoid = SVC(kernel='sigmoid', C=0.02)
svm_sigmoid.fit(train_moons_points, train_moons_labels)
predictions = svm_sigmoid.predict(test_moons_points)

print('F1-score of SVM classifier with sigmoid kernel function on test set: {:.2f}'.format(
    f1_score(y_true=test_moons_labels, y_pred=predictions)))

plot_decision_regs(moons_points, moons_labels, svm_sigmoid)
```

`Conclusion:`
1. Качество классификации значительно улучшилось. В случае RBF ядра f1-score достигло 99%. Для двух остальных ядер, качество незначительно ухудшилось. Самое худшее качество получилось, когда `kernel='sigmoid'`
2. Границы перестали описываться линейной функцией. Вместо этого они преобрели более сложную форму. Классификаторы научились различать данные нелийнейной структуры
3. Использовал дефолтное значение `degree=3.0`. Вместо этого изменял параметр `C`.

## 1.3 Simpler solution (of a kind)
What is we could use Logisitc Regression to successfully solve this task?

Feature generation is a thing to help here. Different techniques of feature generation are used in real life, couple of them will be covered in additional lectures.

In particular case simple `PolynomialFeatures` ([link](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.PolynomialFeatures.html)) are able to save the day.

Generate the set of new features, train LR on it, plot decision regions, calculate metric.

* Compare SVM's results with this solution (quality, borders type)
* What degree of PolynomialFeatures have you used? Compare with same SVM kernel parameter.

```python
from sklearn.preprocessing import PolynomialFeatures

# Сгенерируем полиномиальные признаки
poly = PolynomialFeatures(degree=4)
moons_poly_points = poly.fit_transform(moons_points)

# Разделим данные на train и test
train_moons_points, test_moons_points, train_moons_labels, test_moons_labels = train_test_split(
    moons_poly_points, moons_labels, test_size=0.3, random_state=42)

# Обучим логистическую регрессию
lr = LogisticRegression(C=2)
lr.fit(train_moons_points, train_moons_labels)
predictions = lr.predict(test_moons_points)

print('F1-score of Logistic Regression classifier on test set: {:.2f}'.format(
    f1_score(y_true=test_moons_labels, y_pred=predictions)))
```

```python
plot_decision_regs(moons_points, moons_labels, lr, poly=poly)
```

`Conclusion:`
1. Получил отличный f1-score = 98%. Удалось достигнуть этого за счет добавления полиномиальных фичей. Добавление полиномиальных фичей по факту - это аналог kernel trick. Только вместо того, чтобы считать ядро, мы руками добавляем новые фичи и переходим в пространство более высокой размерности, в котором изначально линейнонеразделимые признаки становятся разделимыми. 
2. Использовал `degree = 4`. Это позволило достичь хорошего качества.

Поработаем для начала с **логистической регрессией**. Будем перебирать по GridSearch-у параметры `penalty` и `C` для поиска лучшего качетсва.

```python
from sklearn.model_selection import GridSearchCV

lr = LogisticRegression(solver='saga', max_iter=10000, multi_class='multinomial', n_jobs=-1)
params = {'penalty': ['l1', 'l2'], 'C': np.linspace(0.1, 3, 30)}
gc = GridSearchCV(estimator=lr, param_grid=params, scoring='accuracy', cv=5)

# Добавим полиномиальных фичей в датасет 
poly = PolynomialFeatures(3)
poly_points = poly.fit_transform(points)

gc.fit(poly_points, labels)
```

```python
print('The best accuracy : {:.3f}'.format(gc.best_score_))
print('The best params: C = {:.1f}, penalty = {:}'.format(gc.best_params_['C'], gc.best_params_['penalty']))
```

```python
# Обучим логистическую регрессию с учетом найденных параметров и нарисуем decisions regions 
lr = LogisticRegression(solver='saga', max_iter=10000, multi_class='multinomial', n_jobs=-1, C=2.3, penalty='l1')

# Разделим выборку на train и test
train_points, test_points, train_labels, test_labels = train_test_split(
    poly_points, labels, test_size=0.3, random_state=42)

lr.fit(train_points, train_labels)
predictions = lr.predict(test_points)
print('F1-score of Logistic Regression classifier on test set: {:.2f}'.format(
    f1_score(y_true=test_labels, y_pred=predictions, average='macro')))
```

```python
plot_decision_regs(points, labels, lr, poly=poly, cmap=ListedColormap(['#e8bfff','#b0b3ff', '#a9d46a', '#fffdcc']))
```

Сделаем тоже самое для **SVM**. Будем перебирать по GridSearch-у следующие параметры: `kernel` и `C`. 

```python
svm = SVC()
params = {'kernel': ['poly', 'rbf', 'sigmoid'], 'C': np.linspace(0.1, 10, 100)}

gc = GridSearchCV(estimator=svm, param_grid=params, scoring='accuracy', cv=5)
gc.fit(points, labels)

print('The best accuracy : {:.3f}'.format(gc.best_score_))
print('The best params: C = {:.1f}, kernel = {:}'.format(gc.best_params_['C'], gc.best_params_['kernel']))
```

```python
# Обучим SVM с учетом найденных параметров и нарисуем decisions regions 
svm = SVC(kernel='rbf', C=2.9)

# Разделим выборку на train и test
train_points, test_points, train_labels, test_labels = train_test_split(
    points, labels, test_size=0.3, random_state=42)

svm.fit(train_points, train_labels)
predictions = svm.predict(test_points)

print('F1-score of SVM classifier on test set: {:.2f}'.format(
    f1_score(y_true=test_labels, y_pred=predictions, average='macro')))
```

```python
plot_decision_regs(points, labels, svm, cmap=ListedColormap(['#e8bfff','#b0b3ff', '#a9d46a', '#fffdcc']))
```

`Conclusion:` Получили примерно одинаковое качество для двух классификаторов: **LR** и **SVM** - 93% и 95% соответственно. В качестве метрики использовали f1-score.
